```python
import os 
import json 
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
```

```python
parent_path = os.getcwd()[:-5]
secret_file_path = os.path.join(parent_path, 'secret.json')
```

```python
with open(secret_file_path, 'r') as json_file:
    secret = json.loads(json_file.read())
```

```python
spotify_api = spotipy.Spotify(
    auth_manager = SpotifyClientCredentials(client_id = secret['id'], client_secret = secret['secret'])
)
```

```python
playlist_id = "10vg1jWK6rXk6xLnB3r10B"
playlist_uri = f"spotify:playlist:{playlist_id}"
```

```python
results = spotify_api.playlist_items(playlist_id=playlist_id)
```
